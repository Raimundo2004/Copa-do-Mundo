#Carregando e analisando a tabela

```python
#Importando e carregando a tabela
import pandas as pd
tabela = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projetos/Bancos de dados/WorldCupMatches.csv')
tabela.head()
```

```python
#Apagar todos os valores nulos da tabela
tabela = tabela.dropna()
```

```python
#Verificando se contm dados duplicados na tabela
tabela.duplicated().sum()
```

```python
#Apagando os dados duplicados
tabela = tabela.drop_duplicates()
```

```python
tabela.duplicated().sum()
```

```python
#Convertendo os dados 'Year'e todos que derivam com os gols para valores inteiros
tabela['Year'] = tabela['Year'].astype(int)
tabela['Away Team Goals'] = tabela['Away Team Goals'].astype(int)
tabela['Home Team Goals'] = tabela['Home Team Goals'].astype(int)
tabela['Half-time Away Goals'] = tabela['Half-time Away Goals'].astype(int)
tabela['Half-time Home Goals'] = tabela['Half-time Home Goals'].astype(int)
```

```python
tabela.info()
```

```python
#Criando uma funo para aplicar na tabela
def resultado(tabela):
  if tabela['Away Team Goals'] > tabela['Home Team Goals']:
    return tabela['Away Team Name']
  elif tabela['Away Team Goals'] < tabela['Home Team Goals']:
    return tabela['Home Team Name']
  else:
    return 'Empate'

#Inserindo a coluna na tabela
tabela['Winner'] = tabela.apply(resultado,axis=1)
```

```python
def resul_casa(tabela):
  if tabela['Home Team Goals'] > tabela['Away Team Goals']:
    return 'Win'
  elif tabela['Home Team Goals'] < tabela['Away Team Goals']:
    return 'Lose'
  else:
    return 'Draw'


def resul_fora(tabela):
  if tabela['Home Team Goals'] < tabela['Away Team Goals']:
    return 'Win'
  elif tabela['Home Team Goals'] > tabela['Away Team Goals']:
    return 'Lose'
  else:
    return 'Draw'

tabela['Resul_Home'] = tabela.apply(resul_casa, axis=1)
tabela['Resul_Away'] = tabela.apply(resul_fora, axis=1)
```

```python
def pontos_casa(tabela):
  if tabela['Resul_Home']=='Win':
    return 3
  elif tabela['Resul_Home']=='Lose':
    return 0
  else:
    return 1

def pontos_fora(tabela):
  if tabela['Resul_Away']=='Win':
    return 3
  elif tabela['Resul_Away']=='Lose':
    return 0
  else:
    return 1

tabela['point_Home'] = tabela.apply(pontos_casa, axis=1)
tabela['point_Away'] = tabela.apply(pontos_fora, axis=1)
```

```python
tabela
```

**Essa tabela ainda se encontra incompleta, pois tem duas alemanhas, uma  s "Germany" e outra  "Germany FR", ento irei deixar somente Alemanha.**

```python
#Trocando as selees de Germany FR para Germany
def alemanha(nome):
  if nome == 'Germany FR':
      return 'Germany'
  return nome

tabela['Home Team Name'] = tabela['Home Team Name'].apply(alemanha)
tabela['Away Team Name'] = tabela['Away Team Name'].apply(alemanha)
```

```python
#Total de gols que cada edio vez
gols_sel_casa = tabela.groupby(['Year'])['Home Team Goals'].sum()
gols_sel_fora = tabela.groupby(['Year'])['Away Team Goals'].sum()
tabela_gols = gols_sel_casa.add(gols_sel_fora, fill_value=0).astype(int).reset_index()
tabela_gols.columns = ['Year', 'Qtd de gols']

plt.title('Linha do tempo de total de gols por edies da copa do mundo')
sns.lineplot(data=tabela_gols, x='Year', y='Qtd de gols')
plt.xticks(tabela_gols['Year'], rotation=85)
plt.yticks(range(0,250,25))
plt.grid()
plt.show()
```

**Com forme o tempo vai passando, h uma certa progresso em nmeros de gols no campeonato, mas tem um ano especfico que houve uma grande ampliao de gols, que foi em 1954,  irei analisar isso o porque dessa grande ampliao.**

```python
tabela['Year'] = tabela['Year'].astype(int)
tabela_gols['Year'] = tabela_gols['Year'].astype(int)
times_total = (pd.concat([tabela[['Year', 'Home Team Name']],tabela[['Year', 'Away Team Name']]]).drop_duplicates().groupby('Year')['Home Team Name'].nunique().reset_index())
times_total.columns = ['Year', 'Qtd de times']
tabela_gols = tabela_gols.merge(times_total, on='Year', how='left')
sns.regplot(data=tabela_gols, x='Year', y='Qtd de times')
plt.xticks(tabela_gols['Year'], rotation=75)
plt.show()
```

```python
fases = tabela.groupby(['Year'])['Stage'].nunique().reset_index()
tabela_gols = tabela_gols.merge(fases, on='Year', how='inner',suffixes=('','_drop')).filter(regex='^(?!.*_drop)')
sns.regplot(data=tabela_gols, x='Year',y='Stage')
plt.xticks(tabela_gols['Year'],rotation=75)
plt.show()
```

```python
print("Correlaes pelo o nmero de gols na Copa")
print(f'Quantida de times: {tabela_gols['Year'].corr(tabela_gols['Qtd de times']):.2f}')
print(f'Quantidade de etapas: {tabela_gols['Qtd de gols'].corr(tabela_gols['Stage']):.2f}')
```
